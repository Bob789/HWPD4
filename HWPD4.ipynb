```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
df = pd.read_csv('mobile.csv')
display(df.head())
display(df.tail())
display(df.info())
```

```python
# display the first name, no need to use applay
# df_names = df['Customer Name'].apply(lambda x: x.split(' ')[0])
df_names = df['Customer Name'].str.split(' ').str[0]
display(df_names.head())
display(df_names.tail())    
```

```python
# display the company name according to the prefix of the phone number
df['Company Name'] = df['Phone Number'].str[:3].replace({'050' : 'Pelephone', '051' : 'Hot Mobile', '052' : 'Cellcom', '053' : 'Golan Telcom', '054': 'Partner'})
                                                                                                                                                                                                                          
display(df.head())
display(df.tail())
```

```python
# display how many subscribers each company has
# display the names company
df_value_counts = df['Company Name'].value_counts()
display(df_value_counts)
df_company_names = df['Company Name'].unique()
display(df_company_names)
```

```python
# status users surfer / talker by using apply function
df['Type User'] = df.apply(lambda row: 'surfer' if row['Internet'] > row['Talked'] else 'talker', axis=1)
display(df.head())
display(df.tail())
```

```python
df = df.drop(columns=['Type User'])
# status users surfer / talker by using vectoriz
def type_user(internet, talked):
    if internet > talked:
        return 'surfer'
    else:
        return 'talker'
df['Type User'] = np.vectorize(type_user)(df['Internet'] , df['Talked'])
display(df.head())
display(df.tail())
```

```python
df = df.drop(columns=['Type User'])
# status users surfer / talker by using np.where
df['Type User'] = np.where(df['Internet'] > df['Talked'], 'surfer', 'talker')
display(df.head())
display(df.tail())
```

```python
# Sort the subscribers by talk time from lowest to highest.
df = df.sort_values(by='Talked', ascending=True, ignore_index=True)
display(df.head())
display(df.tail())
```

```python
# Sort the subscribers by Talk + Internet time from lowest to highest.
df = df.sort_values(by=(['Talked', 'Internet']), ascending=True, ignore_index=True)
display(df)
```

```python
# Print the top 5 talkers.

print('Top 5 talkers:')
display(df.nlargest(5, columns='Talked'))

# print the 3 subscrubwers with lowest internet usage
print('Top 3 lowest internet usage:')   
display(df.nsmallest(3, columns='Internet'))
```

```python
# print the duplicate rows 
df_duplicates = df[df.duplicated()]
df_duplicates_unique = df_duplicates.drop_duplicates()
display(df_duplicates_unique)
```

```python
# Print the names and phone numbers of 3 random subscribers
df_random = df.sample(3)
df_random = df_random[['Customer Name', 'Phone Number']]
print('3 random subscribers:')
display( df_random)
```
